```python
# Q1. 
# In Python 3.X, what are the names and functions of string object types?
# *************************************************************************************************************************

# str, bytes, bytearray are some specialised sequences.
# bytes and bytearray comes under binary sequence types.

# str: It is used for Unicode text, including ASCII
# Example: 
print(str(1))

# bytes: It is used for binary data with absolute byte values
# Exmaple:
print(str(b'Hello world'))

# bytearray: Return a new array of bytes. The bytearray class is a mutable sequence.
# Example:
print(bytearray.fromhex('2Ef0 F1f2'))
```

```python
# Q4. 
# In Python 3.X, what are the key differences between text-mode and binary-mode files?
# *************************************************************************************************************************

# Text-mode files assume their file content is Unicode text (even if it’s all ASCII) and automatically decode when reading.
# With binary-mode files, bytes are transferred to and from the file unchanged.

# The contents of text-mode files are usually represented as str objects in the script.
# The contents of binary files are represented as bytes (or bytearray) objects.

# In Text-mode files, each line of text is terminated with a special character called EOL (End of Line), which is the new 
# line character (‘\n’) in python by default.
# In binary-mode files, there is no terminator for a line and the data is stored after converting it into machine 
# understandable binary language
```
