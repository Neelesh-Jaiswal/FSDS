```python
# Q1. 
# What is the concept of a metaclass?
# *************************************************************************************************************************

# A metaclass is a class of a class that defines how a class behaves. A class is itself an instance of a 
# metaclass.

# The 'type' class is default metaclass which is responsible for making classes
# Metaclass create Classes and Classes creates objects

# Metaclass is responsible for generation of classes, so we can write our own custom metaclasses to modify the way classes 
# are generated by performing extra actions or injecting code.

# Syntax:    
# class1 = Meta(classname, superclasses, attributedict)


PythonClass = type('PythonClass', (), {'start_date': 'August 2018', 'instructor': 'John Doe'} )
print('Attributes values: ',PythonClass.start_date, PythonClass.instructor)
print('PythonClass: ',PythonClass)

# Here 'PythonClass' is a class name and dictionary has the attributes('start_date','instructor') method of the class.

# If we want to do inheritence i.e 'PythonClass' will inherit superclass 'Test'.

class Test():
    pass

test_obj = type('Test', (), {})

derived_class = type('PythonClass', (Test,), {'start_date': 'August 2018', 'instructor': 'John Doe'} )
print('derived_class: ',derived_class)

# CREATING METACLASSES

print('\n'+'*'*127+'\n')

class MyMeta(type):
    pass

class MyClass(metaclass=MyMeta):
    pass

class MySubclass(MyClass):
    pass

print('Type of MyMeta: ',type(MyMeta))
print('Type of MyClass: ',type(MyClass))
print('Type of MySubclass: ',type(MySubclass))
```

```python
# Q2. 
# What is the best way to declare a class's metaclass?
# *************************************************************************************************************************

# we use a keyword argument 'metaclass' in the class header.

# Example:

class MyMeta(type):
    pass

class MyClass(metaclass=MyMeta):
    pass

class MySubclass(MyClass):
    pass

print('Type of MyMeta: ',type(MyMeta))
print('Type of MyClass: ',type(MyClass))
print('Type of MySubclass: ',type(MySubclass))
```
